```python
import porespy as ps
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
import scipy.ndimage as spim
```

```python
path = "output/256x"

img_path = os.listdir(path)


img = cv2.imread(path+"/"+img_path[0], cv2.IMREAD_GRAYSCALE)
img = img/255


thresh = 0.5
im_bw = 1 - cv2.threshold(img, thresh, 1, cv2.THRESH_BINARY)[1]

plt.imshow(im_bw, cmap = "gray")
```

```python
s = [[0,1,0], [1,1,1], [0,1,0]]


labeled_im, n_labels = spim.label(im_bw, structure = s)
plt.imshow(labeled_im);
n_labels
```

```python
props = ps.metrics.regionprops_3D(labeled_im)
```

```python
df = ps.metrics.props_to_DataFrame(props)
df.columns
```

```python
data = ps.metrics.two_point_correlation(img)
plt.plot(data.distance, data.probability, 'r.tw')
```
